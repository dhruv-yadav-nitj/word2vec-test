Dataset Link: [Google News Dataset](https://www.kaggle.com/datasets/adarshsng/googlenewsvectors)

```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format('/kaggle/input/googlenewsvectors/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
model['man'].shape
```

```python
model['man']
```

```python
model.most_similar('india')
```

```python
model.similarity('king', 'queen')
```

```python
model.similarity('man', 'woman')
```

```python
model.similarity('python', 'door')
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar(vec)
```
